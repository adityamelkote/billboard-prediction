```python
import time
import requests
import re
from bs4 import BeautifulSoup

def get_lyrics(url):
    # Set the URL of the song
    url = url

    # Send an HTTP GET request
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML response
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the lyrics element
        lyrics_element = soup.find('div', class_='Lyrics__Container-sc-1ynbvzw-5')

        # Get the lyrics text
        lyrics = lyrics_element.get_text(separator='\n')
        import re
        lyrics = re.sub(r'\[[^]]*\]', ' ', lyrics)
        
        # Return the lyrics
        
        return lyrics
    else:
        print(f'Failed to retrieve data: {response.status_code}')
        return None

# Example usage
df['Lyrics'] = df['Genius Lyrics Link'].apply(lambda x: get_lyrics(x))
time.sleep(0.3)  # Sleep for 0.01 second between requests to avoid overloading the server
```

```python
from pytube import YouTube
import random
import time
import os

def download_audio(url, rank, song):
    try:
        yt = YouTube(url)
        stream = yt.streams.filter(only_audio=True).first()
        filename = f'{song}'
        stream.download(output_path='Audio Files', filename=filename)
    except Exception as e:
        print(f'Error downloading audio for rank {rank}: {e}')

# Example usage
for i, row in df.iterrows():
    url = row['Audio Link']
    rank = row['Rank']
    sleep_time = random.uniform(.5,2)  
    time.sleep(sleep_time)  # Sleep for the specified amount of time
    download_audio(url, rank,df["Song"])
```
